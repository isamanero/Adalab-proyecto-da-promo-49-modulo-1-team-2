```python
import random

conjunto_palabras_secretas = ["adalab", "animo", "python", "crack", "suerte"]
palabra_secreta = random.choice(conjunto_palabras_secretas)

lista_palabra_secreta = list(palabra_secreta) # Aqui convertimos en lista nuestra palabra secreta
lista_tablero = ["*"] * len(palabra_secreta) # Creamos el tablero, seg칰n la longitud de la palabra secreta
print(lista_tablero)

intentos = 0 #Establecemos un contador de intentos

# Empieza el bucle para adivinar la palabra
while intentos < 9: # 9 son los intentos que se emplean para dibujar el ahorcado
    letra = input("Introduce una letra: ").lower() #usamos lower, para coger siempre las letras en min칰sculas

    if letra in lista_palabra_secreta:
        # Si la letra es correcta, reemplaza los '*' en la posici칩n correcta
        for i in range(len(lista_palabra_secreta)):
            if lista_palabra_secreta[i] == letra: #de i, que es nuestra palabra convertida en lista, est치 la letra del input
                lista_tablero[i] = letra #aqui le decimos que entonces, la sustituya en el tablero

        print(f"Letra correcta! Te quedan letras por adivinar, te quedan {9-intentos} intentos: {' '.join(lista_tablero)}") # el m칠todo join es para unir elementos de dos listas
                                                 # se usa el espacio, para separar los *
        
        if ''.join(lista_tablero) == palabra_secreta:
            print("춰Has acertado la palabra!")
            break  # Sale del bucle si se adivin칩 la palabra

    else:
        intentos +=1
        if intentos == 1:
            print(f"La {letra} no est치 en la palabra secreta, te quedan 8 intentos")
            print("=======")    
        if intentos ==2:
            print(f"La {letra} no est치 en la palabra secreta, te quedan 7 intentos")
            print("|    ")
            print("=======")
        if intentos ==3:
            print(f"La {letra} no est치 en la palabra secreta, te quedan 6 intentos")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==4:
            print(f"La {letra} no est치 en la palabra secreta, te quedan 5 intentos")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==5:
            print(f"La {letra} no est치 en la palabra secreta, te quedan 4 intentos")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==6:
            print(f"La {letra} no est치 en la palabra secreta, te quedan 3 intentos")
            print(" *****")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==7:
            print(f"La {letra} no est치 en la palabra secreta, te quedan 2 intentos")
            print(" *****")
            print("|  游 ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==8:
            print(f"La {letra} no est치 en la palabra secreta, te queda 1 intento")
            print(" *****")
            print("|  游 ")
            print("| /|\\ ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==9:
            print(f"La {letra} no est치 en la palabra secreta, HAS PERDIDO, la palabra era {palabra_secreta}")
            print(" *****")
            print("| 游땻 ")
            print("| /|\\ ")
            print("| / \\  ")
            print("|    ")
            print("|    ")
            print("=======")
            break
        

   

```
