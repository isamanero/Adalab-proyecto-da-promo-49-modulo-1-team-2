```python
import random

conjunto_palabras_secretas = ["adalab", "animo", "python", "crack", "suerte"]
palabra_secreta = random.choice(conjunto_palabras_secretas)

lista_palabra_secreta = list(palabra_secreta) # Aqui convertimos en lista nuestra palabra secreta
lista_tablero = ["*"] * len(palabra_secreta) # Creamos el tablero, segÃºn la longitud de la palabra secreta
print(f"ESTE ES TU TABLERO DE JUEGO: ğŸ•¹ï¸ {lista_tablero}")

intentos = 0 #Establecemos un contador de intentos
print("______________________________________________________")
print("Â¡Â¡Bienvenida al juego de AHORCADO ğŸ•¹ï¸!!\n"
    "Si la letra introducida forma parte de la palabra secreta, podrÃ¡s seguir jugando ğŸ™Œ.\n"
    "Si, por el contrario, no es correcta ğŸ˜, se dibujarÃ¡ el ahorcado y se te restarÃ¡ un intento.\n"
    "Con 9 intentos fallidos el ahorcado se habrÃ¡ completado y habrÃ¡s perdido el juego ğŸ˜­.\n"
    "ğŸ€!!Mucha suerte!!ğŸ€ ")

# Empieza el bucle para adivinar la palabra
while intentos < 9: # 9 son los intentos que se emplean para dibujar el ahorcado
    letra = input("Introduce una letra: ").lower() #usamos lower, para coger siempre las letras en minÃºsculas

    if letra in lista_palabra_secreta:
        # Si la letra es correcta, reemplaza los '*' en la posiciÃ³n correcta
        for i in range(len(lista_palabra_secreta)):
            if lista_palabra_secreta[i] == letra: #de i, que es nuestra palabra convertida en lista, estÃ¡ la letra del input
                lista_tablero[i] = letra #aqui le decimos que entonces, la sustituya en el tablero

        print(f"Letra correcta!      â¡ï¸  {' '.join(lista_tablero)}") # el mÃ©todo join es para unir elementos de dos listas
                                                 # se usa el espacio, para separar los *
        
        if ''.join(lista_tablero) == palabra_secreta:
            print("ğŸ‰ğŸ‰ğŸ‰Â¡HAS ACERTADO LA PALABRA! ğŸ‰ğŸ‰ğŸ‰")
            break  # Sale del bucle si se adivinÃ³ la palabra

    else:
        intentos +=1
        if intentos == 1:
            print(f"La {letra} no estÃ¡ en la palabra secreta, te quedan 8 intentos")
            print("       ")
            print("       ")
            print("       ")
            print("       ")
            print("       ")
            print("=======")    
        if intentos ==2:
            print(f"La {letra} no estÃ¡ en la palabra secreta, te quedan 7 intentos")
            print("       ")
            print("       ")
            print("       ")
            print("       ")
            print("|    ")
            print("=======")
        if intentos ==3:
            print(f"La {letra} no estÃ¡ en la palabra secreta, te quedan 6 intentos")
            print("       ")
            print("       ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==4:
            print(f"La {letra} no estÃ¡ en la palabra secreta, te quedan 5 intentos")
            print("       ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==5:
            print(f"La {letra} no estÃ¡ en la palabra secreta, te quedan 4 intentos")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==6:
            print(f"La {letra} no estÃ¡ en la palabra secreta, te quedan 3 intentos")
            print(" *****")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==7:
            print(f"La {letra} no estÃ¡ en la palabra secreta, te quedan 2 intentos")
            print(" *****")
            print("|  ğŸ˜’ ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==8:
            print(f"La {letra} no estÃ¡ en la palabra secreta, te queda 1 intento")
            print(" *****")
            print("|  ğŸ˜’ ")
            print("| /|\\ ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==9:
            print(f"La {letra} no estÃ¡ en la palabra secreta, HAS PERDIDO, la palabra era {palabra_secreta}")
            print(" *****")
            print("| ğŸ˜µ ")
            print("| /|\\ ")
            print("| / \\  ")
            print("|    ")
            print("|    ")
            print("=======")
            break
        

   

```
